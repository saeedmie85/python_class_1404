```python
class IceCream:
    INEVTORY = 0 #class vriable  or static variable
    CAPACITY = 200000 #gram
    def __init__(self, weight, price, quantity):
        self.weight = weight
        self.price = price
        self.quantity = quantity
        IceCream.INEVTORY += quantity * weight
    
    def sell(self,num):
        if num <= self.quantity:
            self.quantity -= num
            IceCream.INEVTORY -= num * self.weight
            return num
        return f'the quantity is {self.quantity}'

    def buy(self, num):
        total_weight = num * self.weight + IceCream.INEVTORY
        if total_weight <= IceCream.CAPACITY:
            self.quantity += num
            IceCream.INEVTORY += num * self.weight
            return 'done' 
        return 'no capacity'

class MusicalIceCream(IceCream):
    def __init__(self, weight, price, quantity, music):
        super().__init__(weight, price, quantity)
        self.music = music
    
    def sell(self, num):
        super().sell(2 * num)
    
    def paly_music(self):
        return f'play {self.music}'
```

```python
class ContactList(list):
    def search_by_name(self,name):
        result = []
        for c in self:
            if name in c.name:
                result.append(c)
        return result

class SendMailMixin:
    def send_mail(self,messege):
        print(f'send mail to {self.email}')

class Contact:
    ALL_CONTACTS = ContactList()
    def __init__(self,name, phone, email):
        self.name = name
        self.phone = phone
        self.email = email
        Contact.ALL_CONTACTS.append(self)

    def call(self):
        print(f'calling to {self.phone}')

    def __repr__(self):
        return f'{self.name}({self.__class__.__name__})'

class Supplier(Contact, SendMailMixin):
    def order(self, goods):
        return f'ordering {goods}'
    
    def send_sms(self, messege):
        return f'sms sending with messege: {messege}'

class Freind(Contact):
    def __init__(self, name,phone, email, address):
        super().__init__(name, phone, email)
        self.address = address
    
    
class visitor(SendMailMixin):
    def __init__(self, ip_address, browser, v_os, email):
        self.ip_address = ip_address
        self.browser = browser
        self.v_os = v_os
        self.email = email
```
